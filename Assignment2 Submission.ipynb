```python
# Q1
def max_cherries(cake):
    max_cherries_anja = 0
    max_cherries_julia = 0
    num_rows = len(cake)
    num_cols = len(cake[0])
    
    # store the counts of counts of cherries in each row and col (row0: 2 cherries, row1: 4 cherries, col0: 2 cherries)
    count_dict = {}
    for i in range(num_rows):
        count_dict['row'+str(i)] = cake[i].count('#')
    for j in range(num_cols):    
        count_dict['col'+str(j)] = [column_wise[j] for column_wise in cake].count('#')
        
#     print(count_dict)
    
    # now cut each row or col and get maximum of first and second part 
    for i in range(num_rows):
        first_half = sum([count_dict['row'+str(k)] for k in range(num_rows) if(k<=i) ]) 
        second_half = sum([count_dict['row'+str(k)] for k in range(num_rows) if(k>i) ]) 
        julia_cherries = max(first_half, second_half) 
        anja_cherries = min(first_half, second_half)        
        max_cherries_anja = max(max_cherries_anja, anja_cherries)
    
    for i in range(num_cols):
        first_half = sum([count_dict['col'+str(k)] for k in range(num_cols) if(k<=i) ]) 
        second_half = sum([count_dict['col'+str(k)] for k in range(num_cols) if(k>i) ])
        julia_cherries = max(first_half, second_half) 
        anja_cherries = min(first_half, second_half)        
        max_cherries_anja = max(max_cherries_anja, anja_cherries)
                
    return max_cherries_anja
    
# Method 2 (prioritising cutting near rows or columns where the number of cherries are highest)
```

```python
# Q2
def spy_locations(heights):
    if( len( heights) > 0 ):
        spy_stack = []
        
        for i in range(len(heights)):
            if(not spy_stack):
                spy_stack.append( (heights[i], i) )
            else:
                if(heights[i] > spy_stack[-1][0]):
                    spy_stack.append( (heights[i], i) )
        
        spy_locations = []
        while(spy_stack):
            spy_locations.append(spy_stack.pop()[1])
        return spy_locations[::-1]
        
    else:
        return []
```

```python
# Q3
def whos_the_winner(s, p):
    sid = s.index(p)
    if( len(s[:sid]) != len(s[sid+ len(p) : ]) ):
        return 'veidt'
    else:
        return 'walter'
```

```python
# Q5
def flip(s_stack, prev_ones_count): 
    i = 0
    while( i < len(s_stack) and s_stack[i] != '1'):
        s_stack[i] = '1'
        prev_ones_count+=1
        i+=1    
    s_stack[i] = '0'
    prev_ones_count-=1
    
    return s_stack, prev_ones_count
    
def get(prev_ones_count):
    return prev_ones_count
    
    
def flip_game(s, queries):
    s_stack = list(s)
    prev_ones_count = s_stack.count('1')
    result_list = []
    
    for k in range(len(queries)):
        if(queries[k] == 'get'):
            result_list.append( get(prev_ones_count) )
        
        else:
            s_stack, prev_ones_count = flip(s_stack, prev_ones_count)
        
    return result_list
```

```python
# Q6
import math
def calculate_points(scoresheet):
    stack = []
    
    op_symbols = {'+':1, '-':2, '*':3, '%': 4, '/': 5}
    
    for i in range(len(scoresheet)):
        if(not stack):
            stack.append( int(scoresheet[i]) )
        
        else:
            if( scoresheet[i] in op_symbols.keys() ):
                num_1 = stack.pop()
                num_2 = stack.pop()
                
                if(scoresheet[i] == '+'):
                    stack.append( int(num_2) )
                    stack.append( int(num_1) )
                    stack.append( int(num_2) + int(num_1) )

                if(scoresheet[i] == '-'):
                    stack.append( int(num_2) )
                    stack.append( int(num_1) )
                    stack.append( int(num_2) - int(num_1) )

                if(scoresheet[i] == '*'):
                    stack.append( int(num_2) )
                    stack.append( int(num_1) )
                    stack.append( int(num_2) * int(num_1) )
                    
                if(scoresheet[i] == '%' and num_1 != 0 and num_2 != 0):
                    stack.append( int(num_2) )
                    stack.append( int(num_1) )
                    stack.append( int(num_2) % int(num_1) )

                if(scoresheet[i] == '/' and num_1 != 0 and num_2 != 0):
                    stack.append( int(num_2) )
                    stack.append( int(num_1) )
                    stack.append( math.floor( int(num_2) / int(num_1) ))
                
            elif(scoresheet[i] == 'I'):
                stack.pop()
            
            elif(scoresheet[i] == 'D'):
                stack.append(stack[-1]*2)
                
            else:
                stack.append( int(scoresheet[i]) )
                
    if(stack):
        return sum(stack)
    else:
        return 0
```

```python
scoresheet = ['5', '2', 'I', 'D', '+']
print(calculate_points(scoresheet))

scoresheet = ['5', '-2', '4', 'I', 'D', '9', '+', '/']
print(calculate_points(scoresheet))

scoresheet = ['30', '-200', '67', 'I', 'D', 'D', '1350', '+', '+', '*', '/', '/']
print(calculate_points(scoresheet))

```

```python
# Q7
class Node :
    def __init__(self, num, next = None):
        self.num = num
        self.next = next
        
class Linked_List():
    
    def __init__(self):
        self.head = None
    
    def insert(self, element, position):
        # create node with the element
        node_to_insert = Node(element)
        count = 0
        
        # insertion at first position
        if(position == 0):
            node_to_insert.next = self.head # add a new node before the current head
            self.head = node_to_insert # replace the old head
            
        # insertion other than first position
        else:
            prev_node = self.head
            while(prev_node != None): # keep searching until element found and position is valid
                if(count == position - 1):
                    break
                prev_node = prev_node.next
                count += 1
            
            if(prev_node != None):
                node_to_insert.next = prev_node.next # make next of current node to next of new node
                prev_node.next = node_to_insert # replace the old node at that postion
                
        return

    def delete(self, element):
        
        # keep searching until element found
        prev_node = self.head
        to_delete = self.head
        count = 0
        
        while(to_delete != None):
            if(to_delete.num == element):
                break
            prev_node = to_delete
            to_delete = to_delete.next
            count += 1
        
        # if deletion is first element
        if(count == 0):
            self.head = to_delete.next
            to_delete = None
            
        
        # if deletion not first element
        else:
            if(to_delete != None):
                prev_node.next = to_delete.next
                to_delete = None
        
        return


    def update(self, search_element, replace_element):
        to_replace = self.head
        
        while(to_replace != None):
            if(to_replace.num == search_element):
                break
            to_replace = to_replace.next
        
        if(to_replace != None):
            to_replace.num = replace_element
            
        return


    def search(self, element):
        to_find = self.head
        
        while(to_find != None):
            if(to_find.num == element):
                return True
            to_find = to_find.next
        
        return False


    def merge_linkedlists(self, linkedlist2):
        # compare both lists by checking current element of one list with the head of another list
        merged_list = Linked_List()
        if(self):
            linkedlist1_head = self.head
        
        linkedlist2_head = linkedlist2.head
        count = 0
        # call next loop with both lists, and one of the lists should have head from the minimum value                     
        while(linkedlist2_head != None or linkedlist2_head != None):
            if(linkedlist1_head == None):
                merged_list.insert(linkedlist2_head.num, count)
                linkedlist2_head = linkedlist2_head.next
            
            elif(linkedlist2_head == None):
                merged_list.insert(linkedlist1_head.num, count)
                linkedlist1_head = linkedlist1_head.next
            
            elif(linkedlist1_head != None and linkedlist2_head != None):
                if(linkedlist1_head.num < linkedlist2_head.num):
#                     print('left', linkedlist1_head.num, linkedlist2_head.num)
                    temp = linkedlist1_head
                    merged_list.insert(temp.num, count)
                    linkedlist1_head = linkedlist1_head.next

                elif(linkedlist1_head.num > linkedlist2_head.num):
#                     print('right', linkedlist1_head.num, linkedlist2_head.num)
                    temp = linkedlist2_head
                    merged_list.insert(temp.num, count)
                    linkedlist2_head = linkedlist2_head.next
            count+=1
            
        return merged_list
    
    def print(self):
        current_node = self.head
        while(current_node != None):
            if(current_node.next != None):
                print(current_node.num, " -> ", end='')
            else:
                print(current_node.num)
            
            current_node = current_node.next
            
```
